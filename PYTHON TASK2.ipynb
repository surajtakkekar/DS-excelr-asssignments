```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy,4io&otracker=nmenu_sub_Electronics_0_Mi'
```

```python
req = requests.get(url)
```

```python
content = BeautifulSoup(req.content,'html.parser')
```

```python
data = content.find_all('div',{'class':'_2kHMtA'})
data
```

```python
links = []
product = []
price=[]
rating=[]
details=[]
start_link ="https://www.flipkart.com"
```

```python
for items in data:
    rest_link =items.find('a')['href']
    name = items.find('div', attrs={'class':'_4rR01T'})
    pr=items.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rt=items.find('div',attrs={'class':'_3LWZlK'})
    dd = items.find('div',attrs={'class':'col col-7-12'})
    product.append(name.text)
    price.append(pr.text)
    rating.append(rt.text)
    details.append(dd.text)
    links.append(start_link+rest_link)                             
                                   
```

```python
print(product[0])
print(price[0])
print(links[0])
print(rating[0])
print(details[0])
```

```python
dataframe = {'product':product, 'price':price, 'Links':links, 'rating':rating, 'details':details}
final_dataframe=pd.DataFrame(dataframe)
print(final_dataframe)
```

```python
final_dataframe.to_csv('fl_task_url.csv')
```
